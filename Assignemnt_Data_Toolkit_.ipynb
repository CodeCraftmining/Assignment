Q1. What is NumPy, and why is it widely used in Python ?

- NumPy (Numerical Python) is a powerful open-source Python library used for numerical and scientific computing. It provides the Ndarray object, which is an efficient and fast multi-dimensional array structure designed for handling large datasets and performing mathematical operations.

- Why is NumPy widely used?
1. Fast Performance:
- NumPy arrays are implemented in C, making operations much faster than Python lists.

2. Vectorized Operations:
- It supports element-wise operations and broadcasting, allowing mathematical computations without using explicit loops.

3. Rich Mathematical Functions:
- NumPy includes built-in functions for linear algebra, statistics, random numbers, Fourier transforms, etc.

4. Foundation for Data Science:
- Libraries like Pandas, Scikit-learn, SciPy, Matplotlib are built on top of NumPy, making it a core tool in data analysis and machine learning.

Q2.How does broadcasting work in NumPy ?

- Broadcasting in NumPy is a technique that allows arithmetic operations on arrays of different shapes by automatically expanding the smaller array to match the larger one. NumPy compares array shapes from the right, and two dimensions are compatible if they are equal or if one of them is 1. If a dimension is 1, NumPy virtually stretches it without copying data. Broadcasting enables operations such as adding a vector to each row of a matrix or adding a scalar to an array. It improves performance by supporting vectorized operations, reducing memory usage, and simplifying code. If the shapes cannot be made compatible, NumPy raises a broadcasting error.

Q4. Explain the use of the groupby() method in Pandas ?

- The groupby() method in Pandas is used to split a DataFrame into groups based on one or more columns and perform operations on each group separately. It follows the Split–Apply–Combine strategy, where the data is first divided into groups, then an aggregation or transformation (such as sum, mean, count, min, max, etc.) is applied to each group, and finally the results are combined into a new DataFrame. It is commonly used for category-wise analysis, data summarization, and statistical computations such as finding average sales per region or total salary per department. The groupby() method makes large datasets easier to analyze and provides powerful tools for grouping, filtering, and transforming structured data.

Q6. What are the differences between NumPy arrays and Python lists) ?

- NumPy arrays differ from Python lists in several ways. NumPy arrays are homogeneous and store elements of the same data type, whereas Python lists can contain mixed data types. NumPy arrays are much more memory-efficient because they use compact, contiguous memory blocks, unlike lists which store references to objects. NumPy arrays offer significantly faster performance due to vectorized operations implemented in C, while Python lists are slower because they rely on Python loops. Arrays support advanced mathematical operations such as elementwise addition, broadcasting, and linear algebra, which lists cannot perform. Additionally, NumPy provides a large collection of optimized numerical functions, making arrays more suitable for scientific computing and data analysis compared to Python lists.

Q8. What does the term “vectorized operation” mean in NumPy ?

- A vectorized operation in NumPy is an operation that is applied to entire arrays or large groups of elements at once, without using explicit Python loops. NumPy performs these operations internally using fast, optimized C-based implementations, which makes them significantly faster and more efficient than traditional looping. Vectorized operations execute elementwise calculations such as addition, multiplication, comparisons, and mathematical functions in a single step. They result in cleaner code, higher performance, and better memory efficiency. Vectorization is one of the main reasons NumPy is widely used for scientific computing and numerical analysis.

Q9. How does Matplotlib differ from Plotly) ?

In contrast, Plotly produces interactive charts with features like zooming, tooltips, panning, and hover information by default, making it suitable for dashboards and web applications. Matplotlib is better for detailed control, scientific plotting, and static reports, whereas Plotly is ideal for interactive data exploration and presentations. Additionally, Matplotlib uses a more low-level approach with extensive customization, while Plotly uses a high-level API that simplifies creating complex, interactive visualizations. Therefore, Matplotlib focuses on static, customizable plotting, while Plotly focuses on interactive, user-friendly, and web-ready visualizations.

Q17. Explain the difference between apply() and map() in Pandas ?

- In Pandas, map() is used to apply a function, dictionary, or mapping to each element of a Series individually, whereas apply() is more flexible and can be used on both Series and DataFrames to apply a function elementwise, row-wise, or column-wise. map() is limited to elementwise transformations on a Series, while apply() can perform complex operations along axes in a DataFrame. map() is generally used for mapping or transforming Series values, whereas apply() is suitable for aggregations, row/column operations, and applying custom functions to both Series and DataFrames.

Q21. Why is NumPy’s array slicing faster than Python’s list slicing ?

- NumPy’s array slicing is faster than Python’s list slicing because NumPy arrays are stored in contiguous memory blocks, allowing direct access via offsets. Slicing in NumPy creates views rather than copies, and operations are executed in optimized C code, avoiding Python-level loops. Unlike Python lists, which store references to objects and require dynamic type checks, NumPy arrays are homogeneous, reducing overhead. Additionally, NumPy uses low-level pointer arithmetic for indexing, making slicing and subarray operations highly efficient. These factors make NumPy slicing significantly faster than Python list slicing, especially for large datasets.

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'A': [1,2,3,4,5],
    'B': [5,4,3,2,1],
    'C': [2,3,4,5,6],
    'D': [5,3,2,4,1]
}
df = pd.DataFrame(data)

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

13. Create a 3D scatter plot using Plotly

```python
import plotly.graph_objects as go


x = [1, 2, 3, 4, 5]
y = [10, 11, 12, 13, 14]
z = [5, 7, 9, 11, 13]


fig = go.Figure(data=[
    go.Scatter3d(
        x=x,
        y=y,
        z=z,
        mode='markers'
    )
])


fig.update_layout(
    title='3D Scatter Plot using Plotly',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)


fig.show()
```
