Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.

-  Simple Linear Regression (SLR) is a statistical and machine learning method used to analyze the relationship between two variables. In this method, one variable is independent (X) and the other is dependent (Y). The goal is to predict the value of the dependent variable using the independent variable.

y = Dependent variable (predicted output)

x = Independent variable (input variable)

a = Intercept (value of y when x = 0)

b = Slope of the line (rate of change in y for one unit change in x)

The main purpose of Simple Linear Regression is to predict future values and understand the relationship between variables. It helps in identifying whether the relationship is positive or negative and how strongly the variables are related.

Question 2: What are the key assumptions of Simple Linear Regression?

- Simple Linear Regression works properly only when certain assumptions are satisfied. These assumptions ensure that the regression model gives accurate and reliable results.

1. Linearity
There must be a linear relationship between the independent variable (X) and the dependent variable (Y). This means the data should follow a straight-line pattern

2. Independence of Errors
The residuals (errors) should be independent of each other. The error of one observation should not influence the error of another observation.

3. Homoscedasticity
The variance of the errors should be constant for all values of X. This means the spread of residuals should remain the same across the regression line.

4. Normality of Errors
The residuals (errors) should be normally distributed. This helps in making valid statistical inferences and hypothesis testing.

5. No Significant Outliers
The dataset should not contain extreme outliers, because they can strongly affect the regression line and produce inaccurate predictions.

Question 5: What is the method of least squares in linear regression ?

- The Method of Least Squares is a mathematical technique used in linear regression to find the best-fitting straight line for a set of data points. The goal is to determine the values of the intercept (a) and slope (b) so that the regression line fits the data as closely as possible.

In real data, the predicted value from the regression line may not exactly match the actual value. The difference between the actual value and predicted value is called the error or residual.

The least squares method works by minimizing the sum of the squares of these errors. Squaring the errors ensures that positive and negative errors do not cancel each other and gives more weight to larger errors

Steps of the Method ->

Calculate the difference between the actual value (Y) and the predicted value (Ŷ).

Square each of these differences.

Add all the squared differences.

Choose the values of a and b that make this total as small as possible.

Question 7: Name and briefly describe three common evaluation metrics for regression

- In regression models, evaluation metrics are used to measure how well the model predicts the actual values. Three common metrics are:

1. Mean Absolute Error (MAE)
Mean Absolute Error measures the average absolute difference between the actual values and predicted values.
It shows how far the predictions are from the real values on average.

Smaller MAE indicates better model performance.

2. Mean Squared Error (MSE)
Mean Squared Error calculates the average of the squared differences between actual and predicted values.
Squaring the errors gives more importance to large errors, which helps detect models with big mistakes.

Lower MSE means the model predictions are more accurate.

3. Root Mean Squared Error (RMSE)
Root Mean Squared Error is the square root of the Mean Squared Error.
It measures the standard deviation of prediction errors and is expressed in the same unit as the target variable.

A smaller RMSE value indicates a better fitting regression model.

Question 8: What is the purpose of the R-squared metric in regression analysis ?

- R-squared (R²) is a statistical metric used to measure how well a regression model fits the data. It represents the proportion of the variance in the dependent variable that is explained by the independent variable(s) in the model.

The value of R² ranges from 0 to 1:

R² = 0 → The model does not explain the variability in the data.

R² = 1 → The model perfectly explains the variability in the data.

Question 9: Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.
(Include your Python code and output in the code box below.)


```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

model = LinearRegression()


model.fit(X, Y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```
